```python
class Car:
  def __init__(self, model_code, model_type, engine_type, colours, tank_capacity, battery_capacity, features):
        self.model_code = model_code
        self.model_type = model_type
        self.engine_type = engine_type
        self.colours = colours if colours else "Not Selected"
        self.tank_capacity = tank_capacity
        self.battery_capacity = battery_capacity
        self.features = features if features else "None"

  def show_info(self):
    print("....Car Manufacturing plant:....")
    print(f"Model code : {self.model_code}")
    print(f"Model type : {self.model_type}")
    print(f"Engine type : {self.engine_type}")
    print(f"Tank capacity : {self.tank_capacity} liter")
    print(f"Battery capacity : {self.battery_capacity} kWh")
    print(f"Colours : {self.colours}")
    print(f"Features : {','.join (self.features) if self.features else None}")

  Available_colours=["Red","Green","Blue","Black","Silver"]
  Available_features=["Ac","Sunroof","Music_player","Airbags"]

class M01(Car):
  def __init__(self,colours):
    super().__init__("M01","Manual","Gasoline",colours,tank_capacity=35,battery_capacity=0,features=['Ac'])

class M02(Car):
  def __init__(self,colours):
    super().__init__("M02","Automatic","Gasoline",colours,tank_capacity=45,battery_capacity=0,features=['Sunroof'])


class M03(Car):
  def __init__(self,colours):
    super().__init__("M03","Automatic","Electric",colours,tank_capacity=0,battery_capacity=45,features=['Ac','Sunroof'])


class M04(Car):
  def __init__(self,colours,engine_type):
    if engine_type=="Gasoline":
      super().__init__("M04","Automatic","Gasoline",colours,tank_capacity=65,battery_capacity=0,features=['Airbags'])

    elif engine_type=="Electric":
      super().__init__("M04","Automatic","Electric",colours,tank_capacity=0,battery_capacity=55,features=['Music_player'])

    elif engine_type=="Hybrid":
      super().__init__("M04","Automatic","Gasoline/Electric",colours,tank_capacity=65,battery_capacity=55,features=['Ac','Airbags','Music_player','Sunroof'])


def user_choice(x,y):
  while True:
    try:
      choice=input(y).strip().capitalize()
      if choice in x:
        return choice

      else:
        print('Oops!invalid input,try Again')

    except Exception as e:
      print('Wrong input ,please selected valid input:',e)

def select_car():
  try:
    model_type=user_choice(['Manual','Automatic'],'choose model_type(Manual/Automatic):')
    if model_type=='Manual':
      print('for "Manual" there is only one model available is "M01":')
      print('For "M01", there is only one engine_type is available :"Gasoline" ')
      engine_use=['Gasoline']
      while True:
          engine_type = user_choice(['Gasoline', 'Electric', 'Hybrid'],
                                          'Choose engine_type (Gasoline/Electric/Hybrid): ')
          if engine_type not in engine_use:
            print(f"For 'M01', there is only one engine_type available is :{engine_use}")
          else:
            print('You choosed right one!')
            break

      colour=user_choice(Car.Available_colours,f"Available colours: {Car.Available_colours}, choose one:")
      return M01(colour)

    elif model_type=='Automatic':
         print("For 'Automatic', available models are ['M02','M03','M04']")
         model_code = user_choice(['M02', 'M03', 'M04'], 'Choose model_code (M02/M03/M04):')
         if model_code=='M02':
          print('for "M02", there is only one engine_type available is:Gasoline')
          engine_use=['Gasoline']
          while True:
            engine_type = user_choice(['Gasoline', 'Electric', 'Hybrid'],
                                          'Choose engine_type (Gasoline/Electric/Hybrid):')
            if engine_type not in engine_use:
              print(f"Oops! you choose wrong , please select :{engine_use}")
            else:
              print('You choosed right one!')
              break

          colour=user_choice(Car.Available_colours,f"Available colours: {Car.Available_colours}, choose one:")
          return M02(colour)

         elif model_code=='M03':
          print('for "M03", there is only one engine_type availabe is :Electric')
          engine_use=['Electric']
          while True:
            engine_type=user_choice(['Gasoline','Electric','Hybrid'],'Choose engine_type(Gasoline/Electric/Hybrid):')

            if engine_type not in engine_use:
              print(f"oops! you choose wrong, please select :{engine_use}")
            else:
              print('you choose right one!')
              break

          colour=user_choice(Car.Available_colours,f"Available colours: {Car.Available_colours}, choose one:")
          return M03(colour)

         elif model_code == 'M04':
          print('For "M04", engine_type you can select Gasoline, Electric, or Hybrid')
          engine_type = user_choice(['Gasoline', 'Electric', 'Hybrid'],'Choose engine_type (Gasoline/Electric/Hybrid):')
          colour = user_choice(Car.Available_colours,f"Available colours: {Car.Available_colours}, choose one:")
          return M04(colour, engine_type)


  except Exception as e:
   print('Error occured',e)

car1=select_car()
car1.show_info()
```
