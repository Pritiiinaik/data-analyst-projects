```python
import random
print(" Welcome to the Number Guessing Game:")
num=random.randint(1,100)
print(num)

guesses=[]

while True:
  user_input=int(input('enter the guess num'))
  guesses.append(user_input)
  if user_input==num:
    print("you guess correct number")
    break
  if user_input<1 or user_input>100:
      print('OUT OF BOUNDS')
      continue
    #first guess of player:
  if len(guesses)==1:
     if abs(user_input-num)<=10:
          print('WARM')
     else:
          print('COLD')
    #compare cuurent num with previous guess::
  else:
    if abs(user_input-num)< abs(guesses[-2]-num):
          print('WARMER')
    else:
          print('COLDER')
```
